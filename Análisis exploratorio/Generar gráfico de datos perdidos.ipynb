```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv("RBE.csv")

# Convertir la columna FECHA_RBE a tipo datetime
df['FECHA_RBE'] = pd.to_datetime(df['FECHA_RBE'])

# Filtrar los datos para la estación 'M_001_A_XV'
df_station = df[df['ID_CIDERH'] == 'M_001_A_XV']

# Filtrar los datos para incluir solo los años desde 1982 hasta 2008
df_selected_years = df_station[(df_station['FECHA_RBE'].dt.year >= 1982) & (df_station['FECHA_RBE'].dt.year <= 2008)]

# Graficar los datos
plt.figure(figsize=(8,4))
plt.plot(df_selected_years['FECHA_RBE'], df_selected_years['PP_RBE'])
plt.title('Precipitation Station M_001_A_XV', fontsize=16, fontweight='bold')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Precipitation (mm)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Leer el nuevo archivo CSV
df_selected_years = pd.read_csv("PP.csv")

# Filtrar los datos para la estación 'M_001_B_II'
df_station = df_selected_years[df_selected_years['ID_CIDERH'] == 'M_001_B_II']

# Convertir la columna FECHA_RBE a tipo datetime
df_station['FECHA_RBE'] = pd.to_datetime(df_station['FECHA_RBE'])

# Reemplazar los valores -99.9 con NaN para que no se consideren en la gráfica
df_station['PP_RBE'].replace(-99.9, np.nan, inplace=True)

# Graficar los datos limpios
plt.figure(figsize=(8, 4))

# Solo se traza la línea donde hay datos válidos (no NaN)
plt.plot(df_station['FECHA_RBE'], df_station['PP_RBE'])

plt.title('Precipitation Station M_001_B_II', fontsize=16, fontweight='bold')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Precipitation (mm)', fontsize=14)
plt.grid(True)

# Establecer el límite inferior del eje x en 1982-01-01
plt.xlim(left=pd.Timestamp('1982-01-01'))

# Establecer el formato del eje x para mostrar solo los años
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator(6))

# Rotar las fechas para que encajen y se lean mejor
plt.gcf().autofmt_xdate()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Leer el archivo CSV
df = pd.read_csv("RBE.csv")

# Convertir la columna FECHA_RBE a tipo datetime
df['FECHA_RBE'] = pd.to_datetime(df['FECHA_RBE'])

# Filtrar los datos para la estación 'M_001_B_II'
df_station = df[df['ID_CIDERH'] == 'M_001_B_II']

# Filtrar los datos para incluir solo los años desde 1993 hasta 2009
df_selected_years = df_station[(df_station['FECHA_RBE'].dt.year >= 1993) & (df_station['FECHA_RBE'].dt.year <= 2009)]

# Reemplazar los valores -99.9 con NaN para que no se consideren en la gráfica
df_selected_years['TEMP_MAX_RBE'].replace(-99.9, np.nan, inplace=True)

# Eliminar filas donde CONF_TEMP_MAX_RBE no tiene datos
df_filtered = df_selected_years[df_selected_years['CONF_TEMP_MAX_RBE'].notna()]

# Graficar los datos
plt.figure(figsize=(8,4))
plt.plot(df_filtered['FECHA_RBE'], df_filtered['TEMP_MAX_RBE'], linestyle='-')

# Establecer el título y las etiquetas del gráfico
plt.title('Maximum Temperature Station M_001_B_II', fontsize=16, fontweight='bold')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Maximum Temperature (°C)', fontsize=14)
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
